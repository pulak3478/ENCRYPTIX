In this project, we used the Titanic dataset to build a logistic regression model to predict whether a passenger survived the tragic sinking of the Titanic. The following key steps were undertaken:

Data Exploration and Preprocessing: We examined the dataset, handled missing values, and encoded categorical variables. Missing values in the Age column were filled with the median age, and the Embarked column was filled with the most common port. The Cabin column, which had a significant amount of missing data, was dropped.

Feature Selection: We selected relevant features for our model, excluding non-informative columns such as Name and Ticket.

Model Training: We split the data into training and testing sets and trained a logistic regression model on the training set.

Model Evaluation: The model was evaluated using the testing set, achieving an accuracy of [77%]. The confusion matrix and classification report provided further insights into the model's performance.
